## Statmech - Plank Distribution
#### 10/8/2024

```python
import numpy as np
import matplotlib.pyplot as plt

# Function if using wavelength
def planck_disto_wavelength(T, wavelength):
    '''
    Calculates the Planck Distribution for a given Temperature and Wavelength.
    Uses SI units for h, c and k_B.
    Give a temperature 'T' in Kelvin, and a wavelength 'wavelength' in meters.
    
    Returns the spectral radiance, in W * m2* sr1 * Hz1
    '''
    # Constants
    h = 6.62607015e-34  # Planck's constant (J s)
    c = 2.998e8         # Speed of light (m/s)
    k_B = 1.380649e-23  # Boltzmann constant (J/K)

    # Calculate energy density using Planck's law
    spectral_radiance = (2 * h * c**2) / (wavelength**5) * (1 / (np.exp(h * c / (wavelength * k_B * T)) - 1))
    
    return spectral_radiance

# Function if using frequency (this is what is used below)
def planck_disto_frequency(T, frequency):
    '''
    Calculates the Planck Distribution for a given Temperature and Frequency.
    Uses SI units for h, c and k_B.
    Give a temperature 'T' in Kelvin, and a frequency 'frequency' in Hz (1/s).
    
    Returns the spectral radiance, in W * m2* sr1 * Hz1
    '''
    # Constants
    h = 6.62607015e-34  # Planck's constant (J s)
    c = 2.998e8         # Speed of light (m/s)
    k_B = 1.380649e-23  # Boltzmann constant (J/K)

    # Calculate energy density using Planck's law
    spectral_radiance = (2 * h * frequency**3) / (c**2) * (1 / (np.exp(h * frequency / (k_B * T)) - 1))
    
    return spectral_radiance

# For a set of given temperatures, find the spectral radiance planck distribution for that Temperature
T_s = [2.7,300,3000,5800,10000,1000000,15000000]  # Temperature values in K
V = 1  # Unit volume

# Loop through the temps and get the Planck Distribution for each
for T in T_s:
    x_s = np.arange(0.1, 10.1, 0.1)  # Frequency values base - will need to multiply by factor
    y_s = np.zeros_like(x_s) # This is the object that will hold the spectral radiance data

    # Need the constants for the analytical solutions included
    h = 6.62607015e-34  # J seconds
    c = 299792458  # m/s
    k = 1.380649E-23  # m2 kg s-2 K-1

    # Calculate the total energy E and total photon number N
    E = (8 * np.pi**5 * k**4) / (15 * c**3 * h**3) * V * T**4
    N = 8 * np.pi * V * ((k * T) / (c * h))**3 * 2.404

    # Do the calculations and populate the return data holder
    for i, x in enumerate(x_s):
        frequency_factor = 1e14  # Look in the range of 1e14 Hz
        frequency = x * frequency_factor
        y_s[i] = planck_disto_frequency(T, frequency)
    # Find the max value in the return values for the numerical max
    max_spectral_radiance = max(y_s)
    max_spectral_radiance_x = x_s[np.where(y_s == max_spectral_radiance)]

    # Plotting the distribution
    fig, ax1 = plt.subplots() # Create the plot object

    # Plot the primary axis, in this case frequency
    ax1.plot(x_s, y_s)
    ax1.set_title(f'Planck Distribution at {T}K')
    ax1.set_xlabel('Frequency [Hz] x10^14')
    ax1.set_ylabel('Spectral Radiance [W * m2 * sr1 * Hz1] x10^-16')

    # Set specific ticks for the primary x-axis
    freq_ticks = np.arange(0, 12, 2)  # Ticks at 0, 2, 4, 6, 8, 10 * 10e14
    ax1.set_xticks(freq_ticks)

    # Put in visual wavelength guide
    red_limit = c/700e-9/frequency_factor
    uv_limit = c/380e-9/frequency_factor
    ax1.axvline(red_limit, color='red', alpha=0.25, linestyle='--')
    ax1.text(red_limit + 0.1, max_spectral_radiance, '700 nm', color='red', alpha=0.7)
    ax1.axvline(uv_limit, color='purple', alpha=0.25, linestyle='--')
    ax1.text(uv_limit + 0.1, max_spectral_radiance, '330 nm', color='purple', alpha=0.7)


    # Create a secondary x-axis for wavelength
    ax2 = ax1.twiny()
    wavelengths = c / (freq_ticks * 1e14)  # Calculate corresponding wavelengths for the tick positions
    ax2.set_xlim(ax1.get_xlim())  # Align limits with primary x-axis
    ax2.set_xticks(freq_ticks)  # Set the same ticks as the primary x-axis
    ax2.set_xticklabels([f'{wavelength*1e9:.0f}' for wavelength in wavelengths])  # Convert to nm
    ax2.set_xlabel('Wavelength [nm]')

    plt.show()

    # Print out calculated and numerically determined properties of a given distribution
    print(f'<E> = {E/1e16} [J]')
    print(f'<N> = {round(N)} [photons]')
    print(f'Check <E>/<N>: {round(E/(N*k*T), 1)} = 2.7\n')
    print(f'P (radiation pressure) = {E/(3*V)} [J/m3]')
    print(f'Peak Frequency (numerically) = {max_spectral_radiance_x[0]}e14 Hz')
    print(f'Peak Frequency (analytically) = {round((2.82*k*T)/(1e14*h),4)}e14 Hz')
    print(f'Peak Wavelength (numerically) = {c/(max_spectral_radiance_x[0]* 1e14)} m')
    print(f'Peak Wavelength (analytically) = {c/(round((2.82*k*T)/(1e14*h),4)*1e14)} m')
    print('\n----\n')
```

```python
from numpy import empty,zeros,max
from pylab import imshow,gray,show

# Constants
M = 20         # Grid squares on a side
V = 1.0         # Voltage at top wall
target = 1e-6   # Target accuracy

# Create arrays to hold potential values
phi = zeros([M+1,M+1],float)
phi[0,:] = V
phiprime = empty([M+1,M+1],float)

# Main loop
delta = 1.0
while delta>target:

    # Calculate new values of the potential
    for i in range(M+1):
        for j in range(M+1):
            if i==0 or i==M or j==0 or j==M:
                phiprime[i,j] = phi[i,j]
            else:
                phiprime[i,j] = (phi[i+1,j] + phi[i-1,j] \
                                 + phi[i,j+1] + phi[i,j-1])/4

    # Calculate maximum difference from old values
    delta = max(abs(phi-phiprime))

    # Swap the two arrays around
    phi,phiprime = phiprime,phi

# Make a plot
imshow(phi)
gray()
show()
```
